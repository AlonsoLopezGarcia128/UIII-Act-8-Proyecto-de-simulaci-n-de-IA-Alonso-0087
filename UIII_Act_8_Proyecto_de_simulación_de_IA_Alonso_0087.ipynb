```python
# --- SIMULADOR DE IA: ASISTENTE TÉCNICO 'FLORETE-BOT 2026' ---

# Inicialización de la variable de control para mantener el menú activo [7, 8]
entrenando = True # Variable booleana que controla el ciclo principal

# Interacción inicial con el usuario para personalización [9, 10]
print("SISTEMA DE INTELIGENCIA ARTIFICIAL PARA ESGRIMA ACTIVADO")
nombre = input("Esgrimista, ingresa tu nombre para el registro: ") # Solicita el nombre del usuario
print(f"Bienvenido al Centro de Alto Rendimiento, {nombre}.") # Muestra saludo personalizado
print() # Genera un espacio visual solicitado

# Inicio del ciclo principal del programa tipo menú [1, 2]
while entrenando == True: # El bucle se repite mientras la condición sea verdadera [1, 8]
    # Visualización de opciones de la IA para el esgrimista [11, 12]
    print("--- MENÚ DE ENTRENAMIENTO Y EQUIPO ---")
    print("1. Análisis de Reglamentación del Florete") # Opción 1: Aritmética y Relacional
    print("2. Diagnóstico de Estilo de Combate")     # Opción 2: Lógica y Cadenas
    print("3. Cálculo de Índice de Rendimiento")     # Opción 3: Aritmética Compleja
    print("4. Predicción de Éxito en Asalto")       # Opción 4: Lógica Combinada
    print("5. Salir del Sistema")                   # Opción de salida

    opcion = input("Seleccione una operación (1-5): ") # Recibe la entrada del esgrimista [9, 10]
    print() # Espaciado solicitado

    # Opción 1: Uso de operadores relacionales y aritméticos [13, 14]
    if opcion == "1": # Compara igualdad para entrar en la opción [5, 13]
        peso = float(input("Ingrese el peso de su florete en gramos: ")) # Lee peso del arma
        # Un florete debe pesar menos de 500g según el reglamento [5, 15]
        if peso > 500: # Operador relacional 'mayor que' [5, 13]
            print("IA ALERTA: El florete excede el peso reglamentario. Reduzca el peso.") # Mensaje de advertencia
        elif peso < 300: # 'elif' para evaluar otro rango de peso [3, 16]
            print("IA REPORTE: El arma es muy ligera. Verifique la estabilidad de la hoja.") # Mensaje informativo
        else: # Se ejecuta si el peso es correcto (entre 300 y 500) [3, 17]
            print("IA REPORTE: Peso óptimo. Florete aprobado para competencia.") # Confirmación técnica

    # Opción 2: Uso de operadores lógicos y comparaciones de texto [5, 15]
    elif opcion == "2": # Evalúa si el usuario eligió la opción 2 [3, 16]
        posicion = input("¿Su guardia es Alta, Media o Baja?: ") # Recibe texto
        movimiento = input("¿Su desplazamiento es Ofensivo o Defensivo?: ") # Recibe segundo parámetro
        # Lógica para determinar el estilo usando 'and' [5, 15]
        if posicion == "Alta" and movimiento == "Ofensivo": # Ambas deben ser ciertas [5, 15]
            print("DIAGNÓSTICO: Tu estilo es el 'Ataque Directo'. Prioriza el fondo.") # Estilo agresivo
        elif posicion == "Baja" or movimiento == "Defensivo": # Basta con que una se cumpla [5, 15]
            print("DIAGNÓSTICO: Tu estilo es el 'Contraataque'. Espera el error del rival.") # Estilo conservador
        else: # Caso para combinaciones no previstas [3, 18]
            print("DIAGNÓSTICO: Estilo equilibrado. Mantén la distancia de seguridad.") # Resultado general

    # Opción 3: Uso de operadores aritméticos para cálculos de desempeño [19, 20]
    elif opcion == "3":
        estocadas = int(input("Número de estocadas intentadas: ")) # Entrada numérica entera
        puntos = int(input("Número de toques logrados: ")) # Entrada numérica entera
        # Cálculo de efectividad usando división y multiplicación [14, 20]
        if estocadas != 0: # Operador relacional 'distinto de' para evitar división por cero [5, 13]
            efectividad = (puntos / estocadas) * 100 # Aritmética de porcentaje [20, 21]
            print(f"IA CÁLCULO: Tu precisión es del {efectividad}% en este asalto.") # Muestra el resultado
        else:
            print("IA ERROR: No se puede calcular sin intentos registrados.") # Manejo de error lógico [22]

    # Opción 4: Lógica compleja con operadores múltiples y booleanos [23, 24]
    elif opcion == "4":
        energia = int(input("Nivel de energía actual (1-10): ")) # Parámetro 1
        rival = int(input("Experiencia del rival (1-10): ")) # Parámetro 2
        # Predicción basada en múltiples condiciones usando paréntesis [25, 26]
        if (energia > 7 and rival < 5) or (energia == 10): # Lógica: superioridad o energía máxima [15, 26]
            print("PREDICCIÓN IA: Alta probabilidad de victoria. ¡Inicia el asalto!") # Predicción positiva
        else:
            print("PREDICCIÓN IA: Riesgo detectado. Analiza los movimientos del oponente.") # Predicción cautelosa

    # Opción 5: Terminación del ciclo [8, 27]
    elif opcion == "5":
        print(f"Cerrando sesión. ¡Buen entrenamiento, {nombre}!") # Despedida interactiva
        entrenando = False # Cambia la variable a False para detener el 'while' [7, 8]

    # Manejo de opciones fuera de rango [28, 29]
    else:
        print("IA AVISO: Opción no reconocida. Por favor, selecciona del 1 al 5.") # Mensaje de control

    print() # Espaciado al final de cada iteración del menú para legibilidad
```
